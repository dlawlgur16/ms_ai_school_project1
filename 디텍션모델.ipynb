# 데이터 업로드

```python
import os
from PIL import Image, UnidentifiedImageError
from azure.cognitiveservices.vision.customvision.training import CustomVisionTrainingClient
from msrest.authentication import ApiKeyCredentials
from azure.cognitiveservices.vision.customvision.training.models import ImageFileCreateEntry, Region

# 이미지 경로 및 데이터프레임 설정
image_folder = r'./원천데이터/abnormal_230303_add/new_TS1'
selected_df = new_TL1_df.head(50000)

image_list = []
for index, row in selected_df.iterrows():
    image_path = os.path.join(image_folder, row['img_name'])
    if os.path.exists(image_path):
        try:
            # 이미지 열어서 크기를 읽음 (파일이 유효하지 않으면 예외 발생)
            with Image.open(image_path) as img:
                width_px, height_px = img.size
        except UnidentifiedImageError:
            print(f"Skipping file (unidentified image): {image_path}")
            continue

        # 이미지 파일의 binary 데이터를 읽음
        with open(image_path, "rb") as f:
            image_data = f.read()

        # 픽셀 좌표를 상대 좌표로 변환 (0~1 사이)
        face_left_rel = row['face_b_box_x'] / width_px
        face_top_rel = row['face_b_box_y'] / height_px
        face_width_rel = row['face_b_box_w'] / width_px
        face_height_rel = row['face_b_box_h'] / height_px

        body_left_rel = row['body_b_box_x'] / width_px
        body_top_rel = row['body_b_box_y'] / height_px
        body_width_rel = row['body_b_box_w'] / width_px
        body_height_rel = row['body_b_box_h'] / height_px

        # ImageFileCreateEntry 생성 (상대 좌표 Region 사용)
        image_list.append(
            ImageFileCreateEntry(
                name=row['img_name'], 
                contents=image_data, 
                regions=[
                    Region(
                        tag_id="e8fdd47c-51c4-4f5d-a8e8-1305f5bcd1c1", 
                        left=face_left_rel, 
                        top=face_top_rel, 
                        width=face_width_rel, 
                        height=face_height_rel
                    ),
                    Region(
                        tag_id="daef73fd-33e2-4050-b8a2-a88b69354a39", 
                        left=body_left_rel, 
                        top=body_top_rel, 
                        width=body_width_rel, 
                        height=body_height_rel
                    )
                ]
            )
        )
```

```python
# API 설정
ENDPOINT = ""
training_key = ""
project_id = ""

credentials = ApiKeyCredentials(in_headers={"Training-key": training_key})
trainer = CustomVisionTrainingClient(ENDPOINT, credentials)
```

## 전부다 업로드

```python
from azure.cognitiveservices.vision.customvision.training.models import ImageFileCreateBatch

def chunk_list(lst, chunk_size):
    for i in range(0, len(lst), chunk_size):
        yield lst[i:i + chunk_size]

batch_size = 64  # Custom Vision의 최대 배치 크기
total_uploaded = 0

for batch_chunk in chunk_list(image_list, batch_size):
    batch = ImageFileCreateBatch(images=batch_chunk)
    upload_result = trainer.create_images_from_files(project_id, batch=batch)
    if not upload_result.is_batch_successful:
        print("일부 이미지 업로드에 실패했습니다.")
        for image in upload_result.images:
            print(f"Failed image: {image.source_url}, Status: {image.status}")
    else:
        print(f"배치 업로드 성공: {len(batch_chunk)}개의 이미지 업로드됨.")
    total_uploaded += len(batch_chunk)

print(f"총 업로드된 이미지 수: {total_uploaded}")
```

# 이미지 데이터 준비

```python
import os
import shutil
import pandas as pd

# 데이터프레임 (filtered_df) 가정
# filtered_df = pd.read_csv("your_filtered_df.csv")  # 필요 시 CSV에서 로드

# 원본 폴더 및 복사할 폴더 경로
source_folder = "./.gitignore/원천데이터/abnormal_230303_add/new_TS1"
target_folder = "./.gitignore/원천데이터/abnormal_230303_add/1000_new_TS1"

# 대상 폴더가 없으면 생성
os.makedirs(target_folder, exist_ok=True)

# 복사할 파일 리스트
image_names = set(filtered_df['img_name'])  # 중복 제거
copied_count = 0

for img_name in image_names:
    source_path = os.path.join(source_folder, img_name)
    target_path = os.path.join(target_folder, img_name)
    
    if os.path.exists(source_path):  # 파일 존재 여부 확인
        shutil.copy2(source_path, target_path)
        copied_count += 1
        print(f"[✓] Copied: {img_name}")
    else:
        print(f"[⚠] File not found: {img_name}")

print(f"\n[✓] 총 {copied_count}개의 파일이 복사 완료되었습니다!")
```

# 데이터준비(crop)

```python
import os
from azure.cognitiveservices.vision.customvision.prediction import CustomVisionPredictionClient
from msrest.authentication import ApiKeyCredentials
from PIL import Image

# === 설정 ===
PREDICTION_KEY = "EAqbiKHt9sYSwFiO7U1NUBngdvnBoJRmd3ZLl0V0XzKkBbzibzPoJQQJ99BBACYeBjFXJ3w3AAAIACOGYNGQ"
PREDICTION_ENDPOINT = "https://6b034cv20250210-prediction.cognitiveservices.azure.com/"
PROJECT_ID = "5cd011c1-e369-47fd-9059-65cc46239b97"
ITERATION_NAME = "Iteration1"

# 입력 및 출력 폴더 경로
INPUT_FOLDER = "./.gitignore/원천데이터/abnormal_230303_add/1000_new_TS1"
OUTPUT_FOLDER = "./.gitignore/원천데이터/abnormal_230303_add/cropped_TS1_body"

# 출력 폴더 생성 (없으면 생성)
os.makedirs(OUTPUT_FOLDER, exist_ok=True)

# === Azure Custom Vision 예측 클라이언트 설정 ===
credentials = ApiKeyCredentials(in_headers={"Prediction-key": PREDICTION_KEY})
predictor = CustomVisionPredictionClient(PREDICTION_ENDPOINT, credentials)

def detect_and_crop_body(image_path, output_folder, threshold=0.5):
    """
    모델을 사용하여 'body' 태그만 감지하고, 크롭하여 저장하는 함수
    """
    with open(image_path, "rb") as image_data:
        results = predictor.detect_image(PROJECT_ID, ITERATION_NAME, image_data)

    # 원본 이미지 로드
    img = Image.open(image_path)
    width, height = img.size

    # 감지된 객체 중 'body'만 크롭
    for prediction in results.predictions:
        if prediction.probability >= threshold and prediction.tag_name == "Body":
            # 바운딩 박스 좌표 변환
            left = int(prediction.bounding_box.left * width)
            top = int(prediction.bounding_box.top * height)
            box_width = int(prediction.bounding_box.width * width)
            box_height = int(prediction.bounding_box.height * height)
            crop_box = (left, top, left + box_width, top + box_height)

            cropped_img = img.crop(crop_box)

            # 원본 파일명 유지
            output_filename = os.path.basename(image_path)  # ex) "image1.jpg"
            output_path = os.path.join(output_folder, output_filename)
            cropped_img.save(output_path)

            print(f"[✓] Cropped body saved to: {output_path}")
            return  # 첫 번째 'body'만 저장하고 종료

def process_images(input_folder, output_folder, threshold=0.5):
    """
    입력 폴더 내 모든 이미지를 처리하는 함수
    """
    for filename in os.listdir(input_folder):
        if filename.lower().endswith((".jpg", ".jpeg", ".png")):
            image_path = os.path.join(input_folder, filename)
            print(f"[*] Processing: {image_path}")
            detect_and_crop_body(image_path, output_folder, threshold)

# === 실행 ===
process_images(INPUT_FOLDER, OUTPUT_FOLDER, threshold=0.9)
print("[✓] 모든 'body' 크롭 이미지 처리가 완료되었습니다.")
```

```python
import os
from PIL import Image, UnidentifiedImageError
from azure.cognitiveservices.vision.customvision.training import CustomVisionTrainingClient
from msrest.authentication import ApiKeyCredentials
from azure.cognitiveservices.vision.customvision.training.models import ImageFileCreateEntry, ImageFileCreateBatch

# === 1. 설정 ===
ENDPOINT = "https://6b034cv20250210.cognitiveservices.azure.com/"
TRAINING_KEY = "bAUggZC5FijhIfLdkUjdOeTr9LWzag6pbF18aGPoFO9Pd0CMO7cwJQQJ99BBACYeBjFXJ3w3AAAJACOGCtyF"
PROJECT_ID = "3e152413-16cb-432f-a410-8a116193865c"

credentials = ApiKeyCredentials(in_headers={"Training-key": TRAINING_KEY})
trainer = CustomVisionTrainingClient(ENDPOINT, credentials)

# === 2. 이미지 폴더 및 데이터프레임 설정 ===
image_folder = r'./.gitignore/원천데이터/abnormal_230303_add/cropped_TS1_body'
selected_df = filtered_df.head(10000)  # 10,000개 데이터 사용

# === 3. 태그 ID 매핑 ===
tag_map = {
    "운전하다": "aaba31cd-7b0b-4606-8409-a125e51ce607",
    "꾸벅꾸벅졸다": "b06b61da-f297-4f41-8b12-d88e09ddf468",
    "무언가를보다": "5d0f4da7-4542-49ce-8804-eb203bf870f8",
    "몸못가누기": "b6fa3239-e988-4467-90cc-05775b3992be",
    "몸을돌리다": "678d68d8-5528-44b3-822a-d681f1d7546b",
    "핸들을흔들다": "a00a24d9-343e-4e40-93f8-0b8545816394",
    "허리굽히다": "17da7f37-6f0c-4e9a-8e36-695f3abc2f60",
    "무언가를마시다": "3582c995-019a-4e25-9e76-64a9c540fcdb",
    "고개를돌리다": "4fd3e836-07e5-4eaf-8f7c-137cb1900e8b",
    "핸드폰귀에대기": "5667255b-2cb9-4b07-8dbb-c6cabf0de46d"
}
```

```python

# === 4. 이미지 리스트 생성 ===
image_list = []
for index, row in selected_df.iterrows():
    image_path = os.path.join(image_folder, row['img_name'])
    
    if os.path.exists(image_path):  # 파일 존재 여부 확인
        try:
            with open(image_path, "rb") as f:
                image_data = f.read()
        except UnidentifiedImageError:
            print(f"[⚠] Skipping file (unidentified image): {image_path}")
            continue

        # 이미지 업로드 정보 추가 (태그 ID 연결)
        if row['action'] in tag_map:  # action 값이 tag_map에 존재하는지 확인
            image_list.append(
                ImageFileCreateEntry(
                    name=row['img_name'], 
                    contents=image_data,
                    tag_ids=[tag_map[row['action']]]  # 태그 ID 추가
                )
            )
        else:
            print(f"[⚠] Warning: Action '{row['action']}' not found in tag_map!")


```

```python
# === 5. 배치 업로드 함수 ===
def chunk_list(lst, chunk_size):
    """리스트를 batch 크기로 나누는 함수"""
    for i in range(0, len(lst), chunk_size):
        yield lst[i:i + chunk_size]

batch_size = 64  # Custom Vision의 최대 배치 크기
total_uploaded = 0

for batch_chunk in chunk_list(image_list, batch_size):
    batch = ImageFileCreateBatch(images=batch_chunk)
    upload_result = trainer.create_images_from_files(PROJECT_ID, batch=batch)

    if not upload_result.is_batch_successful:
        print("[⚠] 일부 이미지 업로드 실패!")
        for image in upload_result.images:
            print(f"[❌] Failed image: {image.source_url}, Status: {image.status}")
    else:
        print(f"[✓] 배치 업로드 성공: {len(batch_chunk)}개의 이미지 업로드됨.")
    
    total_uploaded += len(batch_chunk)

print(f"\n[✅] 총 {total_uploaded}개의 이미지 업로드 완료!")
```
